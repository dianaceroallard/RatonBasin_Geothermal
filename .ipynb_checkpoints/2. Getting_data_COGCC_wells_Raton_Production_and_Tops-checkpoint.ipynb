### COLORADO WELL DATA DOWNLOADING FROM COGCC & FILTERING RATON BASIN WELLS
#### Download production and formation tops. Recommendation runs 1 time and save in csv format.  

##### Some functions adapted from https://sainosmichelle.github.io/elements.html & https://github.com/mlbelobraydi

```python
import pandas as pd
import geopandas as gpd
from io import BytesIO
from zipfile import ZipFile
import matplotlib.pyplot as plt
import seaborn as sns
from COGCCpy import production
from COGCCpy import formation_tops
import requests
```

```python
extract_path = r'../data/outputs'
path ='https://cogcc.state.co.us/documents/data/downloads/gis/WELLS_SHP.ZIP'
```

```python
resp = requests.get(path).content
zipfile = ZipFile(BytesIO(resp))
file_list = zipfile.namelist()
zipfile.extractall(path=extract_path, members=None, pwd=None)
```

```python
file_list
```

```python
for file in file_list:
            if file.endswith('.shp'):
                file_path = extract_path + r'//' + file
                gdf = gpd.read_file(file_path)
                df_CO = pd.DataFrame(gdf.drop(columns='geometry'))
                df_CO['api10'] = '05' + df_CO['API'].astype(str)
```

```python
df_CO.sample(5)       
```

```python
df_CO.to_csv('Colorado_wells_2021.csv')
```

#### Filtering wells by Township and Range for Raton Basin 

```python
twp_list = ['27S','28S','29S','30S','31S', '32S', '33S', '34S', '35S']
range_list = ['64W', '65W', '66W', '67W', '68W', '69W']
temp = df_CO[df_CO['Township'].isin(twp_list)] 
df_Raton = temp[temp['Range'].isin(range_list)] 
len(df_Raton)
```

```python
df_Raton.describe()
```

```python
df_Raton.to_csv('Raton_wells_2021.csv')
```

```python
fig, ax = plt.subplots(figsize=(25,8))
sns.heatmap(df_Raton.isnull(), ax=ax, cmap="magma")
plt.grid()
plt.show()
```

#### Filtering to well at the Raton Basin deeper than 2000ft

```python
df_Raton_deep=df_Raton[df_Raton['Max_MD'].notnull() & (df_Raton['Max_MD'] >2000)]
```

```python
df_Raton_deep.describe()
```

```python
df_Raton_deep.to_csv('Raton_wells_2021_greater2000ft.csv')
```

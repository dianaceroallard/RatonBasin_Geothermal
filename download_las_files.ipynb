```python
import pooch
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

Abrimos el archivo que tiene las url

```python
df = pd.read_csv("./data/Las available.csv")
df
```

Renombramos los document names para eliminar barras y espacios

```python
for i in df.index:
    df.DocName[i] = df.DocName[i].strip().replace("/", "-").replace(" ", "-")
```

Descargamos los archivos y los guardamos dentro de una carpeta downloads

```python
downloaded_files = {}

for i in df.index:
    api = df.api10[i]
    url = df.DocUrl[i]
    docname = df.DocName[i]
    docname = docname.replace("/", "-").replace(" ", "-")
    
    uid = f"{api}-{docname}"
    full_path = pooch.retrieve(url, known_hash=None, fname=f"{uid}.las", path="downloads")
    downloaded_files[uid] = full_path
```

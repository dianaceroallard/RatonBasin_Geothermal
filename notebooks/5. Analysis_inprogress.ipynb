```python
def make_plot_prod (df_prod):
    sns.distplot(df_prod.index,  kde=False, label='Oil Produced')
```

```python
plot_output = widgets.Output()

dropdown_well = widgets.Dropdown(options = df_Raton_prod['API_Label'].unique(), description='Well:', value='05-055-06040-00')

plot_button = widgets.Button(
    description='Show well production',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click to plot well.'
)
        
def on_button_clicked(b):
    plot_output.clear_output()
    with plot_output:
        make_plot_prod(df_Raton_prod[df_Raton_prod['API_Label'] == dropdown_well.value])
        
plot_button.on_click(on_button_clicked)

display(dropdown_well)
display(plot_button)
display(plot_output)
```

```python
def make_logs (log_df):
    
        #make sure logs are sorted by depth
    logs = log_df.sort_values(by='Depth')
    
    ztop=logs.Depth.min(); zbot=logs.Depth.max()
    
    #cluster=np.repeat(np.expand_dims(logs['LITH_LABEL'].values,1), 100, 1)

    fig, axs = plt.subplots(nrows=1, ncols=4, sharey=True, squeeze=True, figsize=(10, 10), gridspec_kw={'wspace': 0.25})
    fig.subplots_adjust(left=0.05, bottom=0.05, right=0.975, top=0.7, wspace=0.2, hspace=0.2)
    axs[0].set_ylabel('Depth (ft)')
    axs[0].invert_yaxis()
    axs[0].get_xaxis().set_visible(False)
     
    
    # first track GR/SP/CALI/BS logs to display
    ax1 = axs[0].twiny()  
    ax1.plot(log_df.Gamma, log_df.Depth, '-', color='#2ea869', linewidth=0.5)
    ax1.set_xlim(0,250) 
    ax1.set_xlabel('GR (API)', color='#2ea869')
    ax1.minorticks_on()
    ax1.tick_params(axis='x', color='#2ea869')
    ax1.spines['top'].set_position(("axes", 1.0)) # Move the last y-axis spine over to the top by 20% of the width of the axes

    ax1 = axs[1].twiny() 
    ax1.plot(log_df.SP, log_df.Depth, '-', color='#0a0a0a', linewidth=0.7)
    ax1.set_xlim(-200,200) 
    ax1.set_xlabel('SP(mV)', color='#0a0a0a')
    ax1.minorticks_on()
    ax1.spines['top'].set_position(('axes', 1.0)) 
    axs[1].get_xaxis().set_visible(False)
    
    ax1 = axs[2].twiny()  
    ax1.plot(log_df.Density, log_df.Depth, '-', color='#ea0606', linewidth=0.5)
    ax1.set_xlim(1.5,3.0) 
    ax1.set_xlabel('Density (g/cm3)', color='#ea0606')
    ax1.minorticks_on()
    ax1.spines['top'].set_position(('axes', 1.0))
    axs[2].get_xaxis().set_visible(False)
    
    ax1 = axs[3].twiny()  
    ax1.plot(log_df.Resistivity, log_df.Depth, '-', color='#d6dbd7', linewidth=0.5)
    ax1.set_xlim(0.2,20000)  
    ax1.set_xlabel('Resistivity (ohm.m)', color='#d6dbd7')
    ax1.set_xscale('log')
    ax1.minorticks_on()
    ax1.spines['top'].set_position(('axes', 1.0))
    axs[3].get_xaxis().set_visible(False)
```

```python
plot_output = widgets.Output()

dropdown_well = widgets.Dropdown(options = df_Raton_logs['UWI'].unique(), description='Well:', value='05 071 08406 00')

plot_button = widgets.Button(
    description='Show well plot',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click to plot well.'
)
        
def on_button_clicked(b):
    plot_output.clear_output()
    with plot_output:
        make_logs(df_Raton_logs[df_Raton_logs['UWI'] == dropdown_well.value])
        
plot_button.on_click(on_button_clicked)

display(dropdown_well)
display(plot_button)
display(plot_output)
```

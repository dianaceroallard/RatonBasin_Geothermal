##### Las available.csv file with download url for each las for wells in Raton Basin

```python
df = pd.read_csv("../data/inputs/Las available.csv")
df
```

```python
len(df.api10.unique())
```

##### Rename documents to avoid spaces and dashes in the name

```python
for i in df.index:
    df.DocName[i] = df.DocName[i].strip().replace("/", "-").replace(" ", "-")
```

##### Download the las files from DocUrl and safe in "downloads" folder

```python
downloaded_files = {}

for i in df.index:
    api = df.api10[i]
    url = df.DocUrl[i]
    docname = df.DocName[i]
    docname = docname.replace("/", "-").replace(" ", "-")
    
    uid = f"{api}-{docname}"
    full_path = pooch.retrieve(url, known_hash=None, fname=f"{uid}.las", path="downloads")
    downloaded_files[uid] = full_path
```
